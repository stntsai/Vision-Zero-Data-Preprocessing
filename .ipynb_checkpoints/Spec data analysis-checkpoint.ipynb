## 3. use only intersection dataset, concatenate two roads and form an intersection column

```python
def get_intersection(df,a,b):
    df['intersection'] = np.where(intersection[a]> intersection[b], 
                    intersection[a].str.replace('[^\w\s]', '').str.strip().str.lower() +' and ' + intersection[b].str.replace('[^\w\s]', '').str.strip().str.lower(),
                    intersection[b].str.replace('[^\w\s]', '').str.strip().str.lower() +' and ' + intersection[a].str.replace('[^\w\s]', '').str.strip().str.lower())
    return df
    
```

```python
import requests
```

```python
# Google Cloud Platform API post request
# API_KEY_1 = 'AIzaSyCejvb57XSyzW5nVdz57CAP7nJVmcfh5-Q'
# API_KEY_2 = 'AIzaSyCh0z19dIzvI-bgjYkoPZwEfJjh_4K9guM'
API_KEY_3 = 'AIzaSyCQiggN7W5yOXjskRo6G_s93tfahAV7IgE'

def extract_lat_long_via_address(address_or_zipcode):
    lat, lng = None, None
    api_key = API_KEY_3
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    place_url = 'https://maps.googleapis.com/maps/api/place/textsearch/json'
#     endpoint = f"{base_url}?address={address_or_zipcode}&key={api_key}"
    endpoint = f"{place_url}?query={address_or_zipcode}&key={api_key}"
    r = requests.get(endpoint)
    if r.status_code not in range(200, 299):
        return None, None
    try:
        results = r.json()['results'][0]
        lat = results['geometry']['location']['lat']
        lng = results['geometry']['location']['lng']
    except:
        pass
    return lat, lng
```

```python
# don't run this until you know what you are doing!!!!!
def get_coor(array):
    total = len(array)
    cnt = 0
    for inter in array:
        cnt += 1
#         if inter in inter_coor_dict:
#             continue
        lat_lng = extract_lat_long_via_address(inter)
        inter_coor_dict[inter] = lat_lng
        print('finished searching for '+str(cnt)+'/'+str(total)+ ' entries, lat_lng = '+ str(lat_lng))
    return 'finished searching for '+str(cnt)+' entries'
```
